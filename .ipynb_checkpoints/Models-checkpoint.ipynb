### Reading in dataframe

```python

tweet_frame = pd.read_csv('snakebite29082017.csv', sep = ',', encoding = 'latin-1')
```

### Saving the list of words for each tweet in a separate list

```python


sentence_list=[] 
for i in range(tweet_frame['cleaned'].shape[0]):
    l = ast.literal_eval(tweet_frame['cleaned'][i])
    sentence_list.append(l) 
```

### Storing the cleaned list of words for each tweet in the corpus into a single list of all words

```python

word_list = [] 
sentences = tweet_frame['cleaned'].tolist()
for i in range(tweet_frame['cleaned'].shape[0]):
    sentence = sentences[i]
    sentence = ast.literal_eval(sentence)
    word_list = word_list + sentence
```

### Computing the most commonly used words 

```python

count_all = Counter()
count_all.update(word_list)
count_all.most_common(10)

# generate a cloud image prioritizing frequency over rank with weight 0.8
#wordcloud = WordCloud(width=800, height=600, relative_scaling=.8)\
#    .generate_from_frequencies(count_all.items())
#wordcloud.to_file("skills-cloud.png")

# Display the image. Remove the lines below if you don't have matplotlib installed 
# and don't want the interactive display. 
#import matplotlib.pyplot as plt

#plt.figure(figsize=(10,10))
#plt.imshow(wordcloud)
#plt.axis("off")
#plt.show()
```

### Training a word2vec model on the tweet corpus.

Input needs to be a nested list, containing each tweet(tokenised) in a separate list for each row

https://www.tensorflow.org/tutorials/word2vec

```python
# Load Google's pre-trained Word2Vec model.
#It includes word vectors for a vocabulary of 3 million words and phrases that they trained 
#on roughly 100 billion words from a Google News dataset. The vector length is 300 features.

#model = gensim.models.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)  

model = gensim.models.Word2Vec(sentence_list, min_count = 3)

model_doc2vec = gensim.models.Doc2Vec(size=300, window=10, min_count=5, workers=11,alpha=0.025, min_alpha=0.025) # use fixed learning rate
```

### Visualising the model using  TSNE

We can visualize the learned vectors by projecting them down to 2 dimensions using for instance something like the t-SNE dimensionality reduction technique. When we inspect these visualizations it becomes apparent that the vectors capture some general, and in fact quite useful, semantic information about words and their relationships to one another. It was very interesting when we first discovered that certain directions in the induced vector space specialize towards certain semantic relationships, e.g. male-female, verb tense and even country-capital relationships between words, as illustrated in the figure below (see also for example Mikolov et al., 2013).

https://www.tensorflow.org/tutorials/word2vec

```python
#model.wv.vocab is a dict of {word: object of numeric vector}
vocab = list(model.wv.vocab)
X = model[vocab]

#time_start = time.time()
tsne = TSNE(n_components=2, random_state =0, perplexity  = 30.0, n_iter = 5000)
X_tsne = tsne.fit_transform(X)

df = pd.concat([pd.DataFrame(X_tsne),
                pd.Series(vocab)],
               axis=1)

df.columns = ['x', 'y', 'word']

def plot_tsne(x, y, word):
    plt.figure(figsize=(15, 10))
    plt.margins(0)
    plt.axis('on')    
    for i, txt in enumerate(word):
        plt.annotate(txt, (x.iloc[i], y.iloc[i]))
    plt.scatter(x, y, color = 'g', cmap='hsv', alpha=0.9, marker='x', s=0.6, lw=1, edgecolor='') # don't use edges
    plt.show()

plot_tsne(df['x'], df['y'], df['word'])
```
