## Visuals in R

### Installing libraries

```r
rm(list=ls())


library(dplyr, quietly = TRUE, warn.conflicts = FALSE)
library(ggplot2,quietly = TRUE, warn.conflicts = FALSE)
library(tidyr, quietly = TRUE, warn.conflicts = FALSE)
library(reshape2, quietly = TRUE, warn.conflicts = FALSE)
library(wordcloud, quietly = TRUE, warn.conflicts = FALSE)
library(RWeka, quietly = TRUE, warn.conflicts = FALSE)
library(tm, quietly = TRUE, warn.conflicts = FALSE)
require(openNLP, quietly = TRUE, warn.conflicts = FALSE) #similar to RWeka, for NLP processing
suppressWarnings(library(openNLP))
library(ggmap, quietly = TRUE, warn.conflicts = FALSE)

## Importing the twitter snakebite csv file 

snakebite<-read.csv ('snakebite05102017.csv')
head(snakebite)
```

### Word Cloud one-gram 

```r
myCorpus <- Corpus(VectorSource(snakebite$Text))

### Custom stopwords from python nltk toolbox
stop = c('i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 'yours', 'yourself',
             'yourselves', 'he', 'him', 'his', 'himself', 'she', 'her', 'hers', 'herself', 'it', 'its', 'itself',
             'they', 'them', 'their', 'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', 'these',
             'those', 'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do',
             'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while',
             'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 'during', 'before',
             'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 'under', 'again',
             'further', 'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each',
             'few', 'more', 'most', 'other', 'some', 'such', 'no', 'nor', 'not', 'only', 'own', 'same', 'so', 'than',
             'too', 'very', 's', 't', 'can', 'will', 'just', 'don', 'should', 'now', 'd', 'll', 'm', 'o', 're', 've',
             'y', 'ain', 'aren', 'couldn', 'didn', 'doesn', 'hadn', 'hasn', 'haven', 'isn', 'ma', 'mightn', 'mustn',
             'needn', 'shan', 'shouldn', 'wasn', 'weren', 'won', 'wouldn')

myCorpus = tm_map(myCorpus, tolower)
myCorpus = tm_map(myCorpus, removePunctuation)
myCorpus = tm_map(myCorpus, removeNumbers)
corpus <- tm_map(myCorpus, stemDocument)
myCorpus = tm_map(myCorpus, removeWords, c(stopwords("english"), stop))

token_delim <- " \\t\\r\\n.!?,;\"()"
bitoken <- NGramTokenizer(myCorpus, Weka_control(min=1,max=1, delimiters = token_delim))
one_word <- data.frame(table(bitoken))
sort_one <- one_word[order(one_word$Freq,decreasing=TRUE),]
head(sort_one, 20)

set.seed(4363)
wordcloud(sort_one$bitoken,sort_one$Freq,random.order=FALSE,min.freq = 8, colors = brewer.pal(8,"Dark2"))

ggplot(head(sort_one,30), aes(reorder(bitoken,Freq), Freq)) +
  geom_bar(stat = "identity") + coord_flip() +
  xlab("one-gram") + ylab("Frequency") +
  ggtitle("Most frequent single gram")
```

### Word Cloud Bi-grams 

http://text-analytics101.rxnlp.com/2014/11/what-are-n-grams.html

```r
myCorpus <- Corpus(VectorSource(snakebite$Text))

myCorpus = tm_map(myCorpus, tolower)
myCorpus = tm_map(myCorpus, removePunctuation)
myCorpus = tm_map(myCorpus, removeNumbers)
myCorpus = tm_map(myCorpus, removeWords, c(stopwords("english"), stop))

token_delim <- " \\t\\r\\n.!?,;\"()"
bitoken <- NGramTokenizer(myCorpus, Weka_control(min=2,max=2, delimiters = token_delim))
two_word <- data.frame(table(bitoken))
sort_two <- two_word[order(two_word$Freq,decreasing=TRUE),]
head(sort_two, 20)

set.seed(4363)
wordcloud(sort_two$bitoken,sort_two$Freq,random.order=FALSE,min.freq = 8, scale=c(1,.9), colors = brewer.pal(8,"Dark2"))

ggplot(head(sort_two,30), aes(reorder(bitoken,Freq), Freq)) +
  geom_bar(stat = "identity") + coord_flip() +
  xlab("Bigrams") + ylab("Frequency") +
  ggtitle("Most frequent bigrams")
```

### Word Cloud Trigrams

```r
myCorpus <- Corpus(VectorSource(snakebite$Text))
myCorpus <- tm_map(myCorpus, removeNumbers) 
myCorpus <- tm_map(myCorpus, stripWhitespace) 
myCorpus = tm_map(myCorpus, tolower)
myCorpus = tm_map(myCorpus, removePunctuation)
myCorpus = tm_map(myCorpus, removeWords, c(stopwords("english"), stop))

token_delim <- " \\t\\r\\n.!?,;\"()"
Trigram <- NGramTokenizer(myCorpus, Weka_control(min=3,max=3, delimiters = token_delim))
Tab_trigram <- data.frame(table(Trigram))
TrigramGrp <- Tab_trigram[order(Tab_trigram$Freq,decreasing=TRUE),]
head(TrigramGrp, 20)

set.seed(4363)
wordcloud(Tab_trigram$Trigram,Tab_trigram$Freq,random.order=FALSE,min.freq = 4,scale=c(0.9,.5), 
            colors = brewer.pal(8,"Dark2"))

ggplot(head(TrigramGrp,30), aes(reorder(Trigram,Freq), Freq)) +
  geom_bar(stat = "identity") + coord_flip() +
  xlab("Tri-gram") + ylab("Frequency") +
  ggtitle("Most frequent Tri-gram")
```

### Geolocation Map

Use this cell, if you want to calculate new latitudes, longitudes from Google API 

```r
locatedUsers <- !is.na(snakebite$Time_zone)  # Keep only users with location info
locations <- geocode(as.character(snakebite$Time_zone[locatedUsers]))  # Use amazing API to guess

worldMap <- map_data("world")  # Easiest way to grab a world map shapefile
zp1 <- ggplot(worldMap)
zp1 <- zp1 + geom_path(aes(x = long, y = lat, group = group),  # Draw map
                       colour = gray(2/3), lwd = 1/3)

zp1 <- zp1 + geom_point(data = locations,  # Add points indicating users
                        aes(x = lon, y = lat),
                       colour = "RED", alpha = 0.7, size = 1.5)
zp1 <- zp1 + coord_equal()  
zp1 <- zp1 + theme_minimal()  # Drop background annotations
zp1 <- zp1 + ggtitle("Map of twitter user  locations")
zp1 <- zp1 + scale_x_continuous( limits = c( -160 , 160 ) )+
             scale_y_continuous( limits = c( -50 , 65 ))

print(zp1)
snakebite$Timezone_lat = locations$lat
snakebite$Timezone_lon = locations$lon
write.csv(snakebite,'snakebite05102017.csv')
```

Use this cell, if you already have a dataframe with latitudes and longitudes 

```r

locations <- snakebite[,c("Location_lat", "Location_lon")]
timezone <- snakebite[,c("Timezone_lat", "Timezone_lon")]
worldMap <- map_data("world")  # Easiest way to grab a world map shapefile
zp1 <- ggplot(worldMap)

zp1 <- zp1 + geom_path(aes(x = long, y = lat, group = group),  # Draw map
                       colour = gray(2/3), lwd = 1/3)

zp1 <- zp1 + geom_point(data = locations,  # Add points indicating users
                        aes(x = Location_lon, y = Location_lat),
                        colour = "RED", alpha = 0.7, size = 1.5)


zp2 <- zp1 + geom_point(data = timezone,  # Add points indicating users
                        aes(x = Timezone_lon, y = Timezone_lat),
                        colour = "RED", alpha = 0.7, size = 1.5)

zp1 <- zp1 + coord_equal()  
zp1 <- zp1 + theme_minimal()  # Drop background annotations
zp1 <- zp1 + ggtitle("Map of twitter user GPS locations")
zp1 <- zp1 + scale_x_continuous( limits = c( -160 , 160 ) )+
             scale_y_continuous( limits = c( -50 , 65 ))


zp2 <- zp2 + coord_equal()  
zp2 <- zp2 + theme_minimal()  # Drop background annotations
zp2 <- zp2 + ggtitle("Map of twitter user time zones")
zp2 <- zp2 + scale_x_continuous( limits = c( -160 , 160 ) )+
             scale_y_continuous( limits = c( -50 , 65 ))

print(zp1)
print(zp2)
```
